# Clustering KMeans

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

```

```python
vinos=pd.read_csv('caracteristicasDeVinos.csv',engine='python')
vinos.info()
```

```python
vinos.head()
```

```python
vinos_variables=vinos.drop(['Vino'], axis=1)
```

```python
vinos_variables.describe()
```

```python
vinos_norm=(vinos_variables-vinos_variables.min())/(vinos_variables.max()-vinos_variables.min())
vinos_norm
```

```python
## APLICANDO K-MEANS
clustering = KMeans(n_clusters=3, max_iter=300) ## creamos el modelo
clustering.fit(vinos_norm) ## aplica el modelo a la base de datos
```

```python
vinos['KMeans_Clusters'] = clustering.labels_  ## los resultados del clustering se guardan en labels dentro del modelo
vinos.head()
```

```python
## Visualizando los cluster que se formaron, aplicando el analisis de componentes principales. para poder darnos una idea
## de como se formaron los clusters

from sklearn.decomposition import PCA

pca = PCA(n_components=2) ## se crea un modelo de pca con dos componentes para poder graficarlo
pca_vinos = pca.fit_transform(vinos_norm) 
pca_vinos_df = pd.DataFrame(data=pca_vinos, columns=['Componente_1','Componente_2'])
pca_nombres_vinos = pd.concat([pca_vinos_df, vinos[['KMeans_Clusters']]], axis=1)
pca_nombres_vinos
```

```python
fig=plt.figure(figsize=(6,6))
ax= fig.add_subplot(1,1,1)
ax.set_xlabel('Componente 1',fontsize = 15)
ax.set_ylabel('Componente 2',fontsize = 15)
ax.set_title('Componentes principales',fontsize=20)
color_theme=np.array(["blue","green","orange"])
ax.scatter(x=pca_nombres_vinos.Componente_1,y=pca_nombres_vinos.Componente_2,c=color_theme[pca_nombres_vinos.KMeans_Clusters]
           ,s=50)
plt.show()
```
